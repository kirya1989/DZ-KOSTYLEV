```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(-20, 20, 500)

def f1(x):
    return 0.5 * x

def f2(x):
    return np.exp(x) 

def f3(x):
    return np.sin(x)

```

```python
plt.title('x/2')
plt.plot(x, f1(x))
plt.savefig('0.5x.png')
plt.show()
plt.title('$e^x$')
plt.plot(x, f2(x))
plt.savefig('ex.png')
plt.show()
plt.title('sin(x)')
plt.plot(x, f3(x))
plt.savefig('sinx.png')
plt.show()
```
